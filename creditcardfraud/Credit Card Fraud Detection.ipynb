## importing the classifiers

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
```

## printing the accuracy and relevant data for both classifier

```python
classifier = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree Classifier": DecisionTreeClassifier()
}

for name, classifying in classifier.items():
    print(f"\n=========={name}===========")
    classifying.fit(X_data_train, y_data_train)
    y_data_pred = classifying.predict(X_data_test)
    print(f"\n Accuaracy: {accuracy_score(y_data_test, y_data_pred)}")
    print(f"\n Precision: {precision_score(y_data_test, y_data_pred)}")
    print(f"\n Recall: {recall_score(y_data_test, y_data_pred)}")
    print(f"\n F1 Score: {f1_score(y_data_test, y_data_pred)}")
```

```python
from imblearn.over_sampling import SMOTE
```

```python
X_res, y_res = SMOTE().fit_resample(X,y)
```

```python
y_res.value_counts()
```

```python
X_data_train, X_data_test, y_data_train, y_data_test = train_test_split(X_res, y_res, test_size = 0.3, random_state = 42)
```

```python
classifier = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree Classifier": DecisionTreeClassifier()
}

for name, classifying in classifier.items():
    print(f"\n=========={name}===========")
    classifying.fit(X_data_train, y_data_train)
    y_data_pred =classifying.predict(X_data_test)
    print(f"\n Accuaracy: {accuracy_score(y_data_test, y_data_pred)}")
    print(f"\n Precision: {precision_score(y_data_test, y_data_pred)}")
    print(f"\n Recall: {recall_score(y_data_test, y_data_pred)}")
    print(f"\n F1 Score: {f1_score(y_data_test, y_data_pred)}")
```
